```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.plot(x, y, marker='o', linestyle='-', color='b')  # Line plot
plt.title('Line Plot Example')  # Title
plt.xlabel('X-axis')            # X-axis label
plt.ylabel('Y-axis')            # Y-axis label
plt.grid(True)                  # Add grid
plt.show()                      # Display the plot
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

plt.bar(categories, values, color='green')  # Bar chart
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
x = [5, 7, 8, 7, 2]
y = [12, 16, 6, 11, 8]

plt.scatter(x, y, color='red', marker='x')  # Scatter plot
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
data = [22, 87, 5, 43, 56, 73, 55, 54, 11, 20, 51, 5, 79, 31]

plt.hist(data, bins=5, color='purple', edgecolor='black')  # Histogram
plt.title('Histogram Example')
plt.xlabel('Range')
plt.ylabel('Frequency')
plt.show()
```

```python
sizes = [30, 45, 15, 10]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)  # Pie chart
plt.title('Pie Chart Example')
plt.show()
```

```python
plt.figure(figsize=(10, 5))  # Set figure size

# Plot 1
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6], color='blue')
plt.title('Plot 1')

# Plot 2
plt.subplot(1, 2, 2)
plt.bar(['A', 'B', 'C'], [10, 20, 30], color='orange')
plt.title('Plot 2')

plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
 # Customizing Plots
# Change Styles:

plt.style.use('ggplot')  # Popular style
#Add Legends:

plt.plot(x, y, label='Line 1')
plt.legend()  # Show legend

#Save Plot as Image:

plt.savefig('plot.png')
```

```python
import seaborn as sns
```

```python
# Load dataset
data = sns.load_dataset('tips')
print(data.head())
```

```python
sns.lmplot(x='total_bill', y='tip', data=data, hue='sex', markers=['o', 'x'])
plt.title('Scatter Plot with Regression Line')
plt.show()

```

```python
# catagorical plot
#Bar Plot:
sns.barplot(x='day', y='total_bill', data=data)
plt.title('Average Bill Amount per Day')
plt.show()

#Count Plot:
sns.countplot(x='sex', data=data)
plt.title('Count of Male vs Female Customers')
plt.show()
```

```python
#Box Plot:
sns.boxplot(x='day', y='total_bill', data=data, hue='sex')
plt.title('Bill Distribution by Day and Gender')
plt.show()

#Violin Plot:
sns.violinplot(x='day', y='total_bill', data=data, hue='sex', split=True)
plt.title('Bill Amount Distribution')
plt.show()
```

```python
#Heatmap:

corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(data, hue='sex')
plt.show()
```

```python
# Customizing Seaborn Plots
sns.barplot(x='day', y='total_bill', data=data)
plt.title('Title Here')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Save plot as image
sns.barplot(x='day', y='total_bill', data=data, palette='pastel')

# Save plot as image
plt.savefig('seaborn_plot.png')
```
